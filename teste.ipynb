```python
import pandas as pd
import numpy as np

comercio_ajustado = comercio.drop(columns=['id', 'control'])

comercio_ajustado = comercio_ajustado.dropna()

comercio_ajustado = comercio_ajustado[(comercio_ajustado != 0).all(axis=1)]

numeric_cols = comercio_ajustado.select_dtypes(include=np.number).columns

for col in numeric_cols:
    comercio_ajustado[col] = comercio_ajustado[col].astype(str)
    comercio_ajustado = comercio_ajustado[~comercio_ajustado[col].str.len().isin([0])]
    comercio_ajustado[col] = pd.to_numeric(comercio_ajustado[col], errors='coerce')
    comercio_ajustado = comercio_ajustado[comercio_ajustado[col] >= 2008]

comercio_ajustado = comercio_ajustado.dropna()

comercio_ajustado.head(5)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import FuncFormatter
import matplotlib.cm as cm

# Função para formatar o eixo Y em milhões
def format_y_axis(value, _):
    return f'{int(value / 1_000_000):,}'.replace(',', '.')  # Divide por 1 milhão e formata

def create_line_plot(df, title, rows, years, file_name):
    # Gerando uma paleta de cores distintas
    num_colors = len(rows)
    colors = cm.get_cmap('tab10', num_colors)  # Usando uma paleta de 10 cores distintas

    plt.figure(figsize=(14, 6))  # Reduzindo o tamanho do gráfico para melhor visualização
    for i, row in enumerate(rows):
        valid_years = [year for year in years if year in df.columns]
        numeric_years = pd.to_numeric(valid_years)
        data_for_plot = pd.to_numeric(df.iloc[row][valid_years], errors='coerce')
        plt.plot(numeric_years, data_for_plot, label=df['Produto'][row], 
                 marker='o', linestyle='-', color=colors(i), markersize=8, linewidth=2)  # Melhorar a visibilidade dos pontos

    # Adiciona título e rótulos com fontes maiores
    plt.title(title, fontsize=16, fontweight='bold', color='black')  # Cor do título alterada para preto
    plt.xlabel('Ano', fontsize=14, fontweight='bold', color='black')
    plt.ylabel('Volume (em milhões)', fontsize=14, fontweight='bold', color='black')

    # Formatação do eixo Y
    ax = plt.gca()  # Obtém o objeto do eixo atual
    ax.yaxis.set_major_formatter(FuncFormatter(format_y_axis))

    # Melhorar a legibilidade dos ticks
    plt.xticks(numeric_years, fontsize=12, rotation=45)
    plt.yticks(fontsize=12)

    # Outras configurações de design
    plt.legend(fontsize=12, title='Produtos', title_fontsize='13', loc='upper left', bbox_to_anchor=(1, 1), frameon=False)  # Legenda fora do gráfico
    plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)  # Grade mais suave e discreta
    plt.tight_layout()  # Ajusta o espaçamento de forma automática para evitar sobreposição

    # Salva o gráfico como PNG
    plt.savefig(file_name, format='png', dpi=300, bbox_inches='tight')

    # Exibe o gráfico
    plt.show()

# Anos de interesse
years = [str(year) for year in range(2008, 2024)]

# Reinicia o índice do DataFrame
comercio_filtrado = comercio_filtrado.reset_index(drop=True)

# Plota os gráficos e salva como PNG
create_line_plot(comercio_filtrado, "VINHO DE MESA", [1, 2, 3], years, 'vinho_de_mesa.png')
create_line_plot(comercio_filtrado, "VINHO FINO DE MESA", [5, 6, 7], years, 'vinho_fino_de_mesa.png')
create_line_plot(comercio_filtrado, "ESPUMANTE", [8], years, 'espumante.png')
create_line_plot(comercio_filtrado, "SUCO DE UVAS", [9], years, 'suco_de_uvas.png')
create_line_plot(comercio_filtrado, 'OUTROS PRODUTOS COMERCIALIZADOS', [10, 11, 12, 13], years, 'outros_produtos_comercializados.png')












```

#**Exportação de Vinhos**

```python
exportacao = pd.read_csv(r'C:\Users\Usuario\Downloads\FIAP-tech-challenge-1\Dataset and Instructions\ExpVinho.csv', sep=";", thousands=".", decimal=",")
pd.set_option('display.max_columns', None)
exportacao.head()
```

```python
"""
Obs: A primeira coluna do dataset original é referente ao Kg. A segunda coluna é referente ao valor.

Exemplo: Coluna 2010 contem o "kg" e a coluna 2010.1 contem o "valor

"""

new_data = {'País': [], 'Ano': [], 'Volume Litro': [], 'Valor': []}


for column in exportacao.columns[2:]:
    year = column.split('.')[0]  # Extrai o ano do nome da coluna
    if '.' not in column:
        year = int(year)
        if year >= 2008:
            new_data['Ano'].extend([int(year)] * len(exportacao))
            new_data['País'].extend(exportacao['País'])
            new_data['Volume Litro'].extend(exportacao[column])

        # Encontra coluna de valor correspondente
            valor_col = column + '.1'
            if valor_col in exportacao.columns:
                new_data['Valor'].extend(exportacao[valor_col])
            else:
                # Se a coluna de valor não existe, adiciona NaN
                new_data['Valor'].extend([np.nan] * len(exportacao))

pd.set_option('display.max_rows', None)
exportacao_tratada = pd.DataFrame(new_data)
exportacao_tratada = exportacao_tratada[~((exportacao_tratada['Volume Litro'] == 0) | (exportacao_tratada['Valor'] == 0) | (exportacao_tratada['Volume Litro'].isna()) | (exportacao_tratada['Valor'].isna()))]
exportacao_tratada_ordenada = exportacao_tratada.sort_values(by='Valor', ascending=False)
exportacao_tratada_ordenada.head(25)
```

```python
country_counts_exportacao = exportacao_tratada['País'].value_counts()
print("\nContagem de países na base de importação:\n", country_counts_exportacao)
```

```python
paises_frequentes = country_counts_exportacao[country_counts_exportacao > 12].index

# Filtra a tabela 'exportacao_tratada_ordenada' para incluir apenas os países frequentes
tabela_filtrada = exportacao_tratada_ordenada[exportacao_tratada_ordenada['País'].isin(paises_frequentes)]

# Retirando os dados do País "Paraguai" devido caracterizar um Outlier
tabela_filtrada = tabela_filtrada[tabela_filtrada['País']!='Paraguai']

# Agrupa por país e soma os valores de 'Kg' e 'Valor'
tabela_consolidada_exp = tabela_filtrada.groupby('País').agg({'Volume Litro': 'sum', 'Valor': 'sum'})


tabela_consolidada_exp
```

```python
tabela_consolidada_exp['Valor Litro'] = tabela_consolidada_exp['Valor'] / tabela_consolidada_exp['Volume Litro']
```

```python
tabela_consolidada_exp.head()
```

```python
total_valor = tabela_consolidada_exp['Valor'].sum()
total_valor_volume_litro = tabela_consolidada_exp['Volume Litro'].sum()
```

```python
tabela_consolidada_exp.loc['Total'] = [total_valor_volume_litro, total_valor,'']
```

```python
# Substitui " " e "" por NaN nas colunas usando map
tabela_consolidada_exp['Valor'] = tabela_consolidada_exp['Valor'].map(lambda x: np.nan if x in [" ", ""] else x)
tabela_consolidada_exp['Valor Litro'] = tabela_consolidada_exp['Valor Litro'].map(lambda x: np.nan if x in [" ", ""] else x)


# Converte as colunas para float, ignorando os NaNs
tabela_consolidada_exp[['Valor', 'Valor Litro']] = tabela_consolidada_exp[['Valor', 'Valor Litro']].astype(float)
```

```python
linhas = len(tabela_consolidada_exp)

tabela_consolidada_exp.head(linhas)
```

```python
tabela_consolidada_exp_org = tabela_consolidada_exp

# Formatando as colunas 'Valor' e 'Valor Litro' no padrão americano
tabela_consolidada_exp_org[['Valor', 'Valor Litro']] = tabela_consolidada_exp_org[['Valor', 'Valor Litro']].applymap(
    lambda x: "US$ {:,.2f}".format(x) if pd.notnull(x) else ''
)

# Formatando a coluna 'Volume Litro' no padrão americano
tabela_consolidada_exp_org['Volume Litro'] = tabela_consolidada_exp_org['Volume Litro'].apply(
    lambda x: '{:,.2f}'.format(x) if pd.notnull(x) else ''
)

tabela_consolidada_exp_org
```

```python
df_exp = pd.DataFrame({
    'País de origem': ['Brasil'] * len(tabela_consolidada_exp_org),
    'País de destino': tabela_consolidada_exp_org.index,
    'Volume Litro': tabela_consolidada_exp_org['Volume Litro'],
    'Valor Litro': tabela_consolidada_exp_org['Valor Litro'],
    'Valor Total': tabela_consolidada_exp_org['Valor']
})

df_exp = df_exp.reset_index(drop=True)
```
